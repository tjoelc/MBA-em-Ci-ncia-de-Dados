```python
#importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerar dados fictícios de consumo de energia em diferentes bairros ao longo de um ano

np.random.seed(0) # random.seed permite definir o valor inicial para a geração dos números aleatórios

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
bairros = ['Centro', 'Zona Norte', 'Zona Sul', 'Zona Leste', 'Zona Oeste']
dados = pd.DataFrame(np.random.randint(100, 1000, size=(12, 5)), columns=bairros, index=meses)
```

```python
# Análise Exploratória de Dados
# Histograma do consumo de energia ao longo do ano para cada bairro
dados.hist(bins=10, figsize=(10, 6), color='skyblue')
plt.suptitle('Distribuição de Consumo de Energia por Bairro ao Longo do Ano', fontsize=16)
plt.xlabel('Consumo de Energia')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

Na distribuição observamos que o consumo da zona leste por ser menor e na faixa do 200 a 400 kwats/h devido a maior concentração de residencias populares, observamos que as ditribuições de 600kwats/h a 1000kwats/h consumidas na zona sul ,norte e oeste devido concentração de residencias nobres e industrias.

```python
# Boxplot do consumo de energia por bairro
dados.boxplot(figsize=(10, 6), grid=False)
plt.title('Boxplot do Consumo de Energia por Bairro')
plt.ylabel('Consumo de Energia')
plt.show()

```

obs: o bairro que menos consome em média é a zona leste, e zona sul se destaca como maior consumidora de energia.

```python
import matplotlib.pyplot as plt
import numpy as np

# Nomes dos bairros
nomes_bairros = ['Bairro A', 'Bairro B', 'Bairro C']

# Nomes dos meses
nomes_meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
```

```python
# Dados de exemplo - consumo de energia por bairro ao longo do ano
dados = np.random.rand(12, 3) * 1000  # 12 meses, 3 bairros

# Calcular o consumo médio de energia por bairro ao longo do ano
consumo_medio_por_bairro = dados.mean(axis=0)

# Identificar os 3 bairros com os maiores consumos médios de energia
top_bairros = np.argsort(consumo_medio_por_bairro)[-3:]
```

```python
# Selecionar apenas os dados dos 3 bairros com maiores consumos médios
dados_top_bairros = dados[:, top_bairros]

# Plotar mapa de calor dos 3 bairros com maiores consumos médios de energia
plt.figure(figsize=(10, 6))
plt.imshow(dados_top_bairros, cmap='YlGnBu', aspect='auto')
plt.colorbar(label='Consumo de Energia')
plt.title('Mapa de Calor dos Bairros com Maiores Consumos Médios de Energia')
plt.xlabel('Bairros')
plt.ylabel('Meses')
# Adicionar nomes dos bairros no eixo x
plt.xticks(np.arange(len(nomes_bairros)), nomes_bairros)

# Adicionar nomes dos meses no eixo y
plt.yticks(np.arange(len(nomes_meses)), nomes_meses)

plt.show()
```

Aqui observamos o bairros que consomem mais energia e em quais meses do ano.
no bairro A temos em julho o mes com maior consumo.
no bairro B um consumo elevado ocorre no mes de DEZEMBRO
no bairro C temos os meses de Fevereiro como pico de consumo, mas também o me de dezembro
