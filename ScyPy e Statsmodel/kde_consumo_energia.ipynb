# aplicar kde com scipy e statsmodels

```python
# importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import gaussian_kde
import seaborn as sns
```

```python
# leitura dos dados
data = pd.read_csv('consumo_energia.csv')
```

```python
# aplicando função staticsmodel 
kde_sm = sm.nonparametric.KDEUnivariate(data['Usage_kWh'])
kde_sm.fit(bw='normal_reference')
```

```python
# aplicação com seaborn no histplot

sns.histplot(data=data,x=data['Usage_kWh'],kde=True)
```

```python
#criando os dados e ajustando modelo no grafico
kde = gaussian_kde(data['Usage_kWh'])
x_vals = np.linspace(min(data['Usage_kWh']),max(data['Usage_kWh']),100)
plt.plot(x_vals,kde(x_vals),label='KDE')
plt.title('Kernel DEnsity Estimation')
plt.xlabel('Consumo de energia')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
```
