```python
import pandas as pd
```

```python
df = pd.read_excel('corrigindoVendas.xlsx',sheet_name='Planilha1') #importando o arquivo
```

```python
df
```

```python
# arquivo esta desorganizado com falta de 
# colunas e alinhado valores em linhas
```

```python
df_corrigido = pd.melt(df, id_vars='Data',var_name='Sigla',
                       value_name='Venda')
#essa função cria a coluna Data,Sigla e Venda ,
#adicionando as linhas distribuidas nela
```

```python
df_corrigido
```

```python
# convertendo os valores da coluna venda que possui traço para nan
```

```python
df_corrigido['Venda']=pd.to_numeric(df_corrigido['Venda'])
df_corrigido
```

```python
nun = df_corrigido['Venda'].isna()
nun
# verificando se existe nan ainda no dataframe
```

```python
df_corrigido['ano'] = df_corrigido['Data'].dt.year
df_corrigido
#acrescentando coluna ano para analises futuras
```

```python
df2 = pd.read_excel('frete.xlsx')
```

```python
df2
```

```python
# colocando a coluna frete fazendo o merge com a busca pela coluna sigla 

df_procv = pd.merge(df_corrigido,df2[['Sigla','Região','Frete']],on='Sigla',how='left')
```

```python
#chamando o novo df criado
df_procv
```

```python
# agora incluindo a coluna total venda + frete
```

```python
df_procv.assign(Total=df_procv['Venda']+df_procv['Frete'])
```
